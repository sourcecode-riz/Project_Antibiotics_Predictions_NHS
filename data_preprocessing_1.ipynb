### Handling nan

```python
df = df.copy()

df.loc[:, 'ADDRESS_1'].fillna('UNKNOWN', inplace=True)
df.loc[:, 'ADDRESS_2'].fillna('UNKNOWN', inplace=True)
df.loc[:, 'ADDRESS_3'].fillna('UNKNOWN', inplace=True)
df.loc[:, 'ADDRESS_4'].fillna('UNKNOWN', inplace=True)
df.loc[:, 'POSTCODE'].fillna('UNKNOWN', inplace=True)

df['ADDRESS_1'] = df['ADDRESS_1'].fillna('UNKNOWN')
df['ADDRESS_2'] = df['ADDRESS_2'].fillna('UNKNOWN')
df['ADDRESS_3'] = df['ADDRESS_3'].fillna('UNKNOWN')
df['ADDRESS_4'] = df['ADDRESS_4'].fillna('UNKNOWN')
df['POSTCODE'] = df['POSTCODE'].fillna('UNKNOWN')



```

### Filtering Antibiotic Data

```python
antib_df = df[df['BNF_DESCRIPTION'].str.contains('antibiotic', case=False, na=False)]
```

### Antibiotics Only Dataframe After Filteration
