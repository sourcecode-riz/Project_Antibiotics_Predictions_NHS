# Merged all dataframes of All 6 csv Files into One File

```python
import pandas as pd

df1 = pd.read_csv('antibiotics_data.csv')
df2 = pd.read_csv('antibiotics_data_2.csv')
df3 = pd.read_csv('antibiotics_data_3.csv')
df4 = pd.read_csv('antibiotics_data_4.csv')
df5 = pd.read_csv('antibiotics_data_5.csv')
df6 = pd.read_csv('antibiotics_data_6.csv')

# Concatenate the dataframes
df = pd.concat([df1, df2, df3, df4, df5, df6])


#Save the merged dataframe to a new CSV file
df.to_csv('df.csv', index=False)

# Display the merged dataframe
df.head()
```
